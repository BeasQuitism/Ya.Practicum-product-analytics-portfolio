#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.2.** Анализ воронок событий.
- Постройте классическую и пошаговую воронки;
- Охарактеризуйте пользовательский путь в приложении и определите узкие места воронки. 


```python
# Проверяем уникальные названия событий
print(hist_df['event_name'].unique())
```

```python
# Рассчитываем количество уникальных пользователей на каждом этапе воронки

full_funnel_steps = [
    'install / open_web',
    'introduction',
    'registration',
    'main_page',
    'onboarding_complete',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

# Считаем уникальных пользователей на каждом шаге
full_funnel_data = {}
for step in full_funnel_steps:
    full_funnel_data[step] = hist_df[hist_df['event_name'] == step]['user_id'].nunique()


print("Сводка по воронке (абсолютные числа)")
for i, (step, count) in enumerate(full_funnel_data.items(), 1):
    print(f"{step}: {count:,} пользователей")
```

```python
# Классическая воронка

fig, ax = plt.subplots(figsize=(10, 6))

steps = list(full_funnel_data.keys())
values = list(full_funnel_data.values())
colors = plt.cm.Blues(np.linspace(0.7, 0.2, len(steps)))

bars = ax.barh(steps, values, color=colors, edgecolor='black')

ax.set_title('Классическая воронка пользователей SollmaFin', fontsize=14, fontweight='bold')
ax.set_xlabel('Количество уникальных пользователей')
ax.invert_yaxis()

# Добавляем подписи
for i, bar in enumerate(bars):
    width = bar.get_width()
    n_users = values[i]
    conv_pct = (n_users / values[0]) * 100
    
    # Количество пользователей
    ax.text(width + (values[0] * 0.01), bar.get_y() + bar.get_height()/2, 
            f'{int(n_users):,}', ha='left', va='center', fontsize=9)
    
    # Процент от старта
    ax.text(width * 0.98, bar.get_y() + bar.get_height()/2, 
            f'{conv_pct:.1f}%', ha='right', va='center', color='black', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.show()
```

```python
# Пошаговая конверсия (Step-by-step)

step_conversions = []
prev_step = None
prev_count = None

for step in full_funnel_steps:
    current_count = full_funnel_data[step]
    
    if prev_step is not None:
        # Конверсия от предыдущего шага
        step_conv = (current_count / prev_count) * 100 if prev_count > 0 else 0
        step_conversions.append({
            'from': prev_step,
            'to': step,
            'conversion': step_conv,
            'dropoff': 100 - step_conv
        })
        print(f"{prev_step} → {step}: {step_conv:.1f}% ({step_conv:.1f}%)")
    
    prev_step = step
    prev_count = current_count
```

```python
plt.figure(figsize=(14, 6))

# Подготовка данных
x_labels = [step['to'] for step in step_conversions]
conversions = [step['conversion'] for step in step_conversions]

plt.bar(x_labels, conversions, edgecolor='black')

plt.title('Пошаговая конверсия между этапами воронки', fontsize=14, fontweight='bold')
plt.ylabel('Конверсия, %')

# Подписи значений
for i, conv in enumerate(conversions):
    plt.text(i, conv + 2, f'{conv:.1f}%', ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```

**Задача 2.3.** Влияние уровня риска актива на открытие второго депозита.
- Изучите, как уровень риска первого актива влияет на конверсию во второй депозит.  

```python
# Фильтруем события покупки актива
asset_purchases = hist_df[hist_df['event_name'] == 'asset_purchase'].copy()

print(f"Всего событий покупки актива: {len(asset_purchases)}")
print(f"Уникальных пользователей с покупками: {asset_purchases['user_id'].nunique()}")

# Находим первую покупку актива для каждого пользователя
# Сортируем по пользователю и времени события
asset_purchases = asset_purchases.sort_values(['user_id', 'event_ts'])

# Берём первую запись для каждого user_id
first_asset = asset_purchases.groupby('user_id').first().reset_index()

# Оставляем нужные колонки: user_id, risk_level, asset, event_ts
first_asset = first_asset[['user_id', 'risk_level', 'asset', 'event_ts']].copy()
first_asset.rename(columns={'event_ts': 'first_asset_ts'}, inplace=True)

print(f"\nПользователей с первой покупкой актива: {len(first_asset)}")
print("\nРаспределение первых покупок по уровню риска:")
print(first_asset['risk_level'].value_counts())
```

```python
# Определяем, кто из этих пользователей открыл второй депозит
users_with_second_deposit = set(
    hist_df[hist_df['event_name'] == 'second_deposit']['user_id'].unique()
)

# Добавляем флаг второго депозита
first_asset['has_second_deposit'] = first_asset['user_id'].isin(users_with_second_deposit).astype(int)

print(f"\nИз пользователей с первой покупкой актива:")
print(f"Открыли второй депозит: {first_asset['has_second_deposit'].sum()} ({first_asset['has_second_deposit'].mean()*100:.1f}%)")
```

```python
# Группируем по risk_level и считаем конверсию
conversion_by_risk = first_asset.groupby('risk_level').agg(
    total_users=('user_id', 'count'),
    with_second_deposit=('has_second_deposit', 'sum')
).reset_index()

# Считаем конверсию в процентах
conversion_by_risk['conversion_rate'] = (
    conversion_by_risk['with_second_deposit'] / conversion_by_risk['total_users'] * 100
)

# Сортируем по уровню риска
risk_order = ['low', 'medium', 'high']
conversion_by_risk['risk_order'] = conversion_by_risk['risk_level'].apply(
    lambda x: risk_order.index(x) if x in risk_order else 0
)
conversion_by_risk = conversion_by_risk.sort_values('risk_order')

print("\nКонверсия во второй депозит по уровню риска первого актива\n")
print(conversion_by_risk[['risk_level', 'total_users', 'with_second_deposit', 'conversion_rate']].to_string(index=False))
```

```python
# Функция для z-теста
def proportions_ztest(success1, nobs1, success2, nobs2):
    p1 = success1 / nobs1
    p2 = success2 / nobs2
    p_pool = (success1 + success2) / (nobs1 + nobs2)
    se = np.sqrt(p_pool * (1 - p_pool) * (1/nobs1 + 1/nobs2))
    z_stat = (p1 - p2) / se
    p_value = 2 * norm.sf(abs(z_stat))
    return z_stat, p_value

results = []
for risk in ['low', 'medium', 'high']:
    risk_data = conversion_matrix[conversion_matrix['risk_level'] == risk]
    
    ctrl_row = risk_data[risk_data['group'] == 'control'].iloc[0]
    test_row = risk_data[risk_data['group'] == 'test'].iloc[0]
    
    n_ctrl = int(ctrl_row['total_users'])
    n_test = int(test_row['total_users'])
    
    success_ctrl = int(ctrl_row['with_second_deposit'])
    success_test = int(test_row['with_second_deposit'])
    
    rate_ctrl = success_ctrl / n_ctrl * 100 if n_ctrl > 0 else 0
    rate_test = success_test / n_test * 100 if n_test > 0 else 0
    
    change = rate_test - rate_ctrl
    
    z_stat, p_value = proportions_ztest(success_test, n_test, success_ctrl, n_ctrl)
    significant = p_value < 0.05
    
    results.append({
        'risk_level': risk, 'rate_control': rate_ctrl, 'rate_test': rate_test, 
        'change': change, 'p_value': p_value, 'significant': significant
    })


results_df = pd.DataFrame(results)
print(results_df.round(3).to_string(index=False))
```

**Задача 3.4.** Анализ метрик А/В-эксперимента. Проанализируйте такие метрики:
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Для каждой метрики выведите: 
- значения в контрольной и тестовой группах, 
- абсолютные и относительные изменения значений в тестовой группе по сравнению с контрольной.

После этого проверьте статистическую значимость различий метрик между группами эксперимента. 

Для ключевой метрики также изучите:
- накопленную динамику изменения по дням эксперимента для каждой группы, 
- стабильность p-value во время эксперимента.

Сделайте промежуточный вывод. 


```python
# Расчет метрик

def calculate_ab_metrics(df, group_name):
    group_data = df[df['group'] == group_name].copy()
    
    # Базовые счётчики
    all_users = group_data['user_id'].nunique()
    
    # Пользователи с депозитами
    first_deposit_users = group_data[group_data['event_name'] == 'first_deposit']['user_id'].nunique()
    second_deposit_users = group_data[group_data['event_name'] == 'second_deposit']['user_id'].nunique()
    paying_users = group_data[group_data['event_name'].isin(['first_deposit', 'second_deposit'])]['user_id'].nunique()
    
    # Суммы депозитов
    deposits = group_data[group_data['event_name'].isin(['first_deposit', 'second_deposit'])]
    total_deposit_sum = deposits['amount'].sum()
    
    # Ключевая метрика: средняя сумма на пользователя
    # Создаём таблицу: каждый пользователь и его общая сумма депозитов
    user_totals = deposits.groupby('user_id')['amount'].sum().reset_index()
    all_users_df = pd.DataFrame({'user_id': group_data['user_id'].unique()})
    all_users_df = all_users_df.merge(user_totals, on='user_id', how='left')
    all_users_df['amount'] = all_users_df['amount'].fillna(0)
    
    key_metric = all_users_df['amount'].mean()
    
    # Барьерная метрика: конверсия в первый депозит
    barrier_metric = first_deposit_users / all_users * 100 if all_users > 0 else 0
    
    # Вспомогательная 1: конверсия 1→2 депозит
    app_metric_1 = second_deposit_users / first_deposit_users * 100 if first_deposit_users > 0 else 0
    
    # Вспомогательная 2: средняя сумма на платящего
    app_metric_2 = total_deposit_sum / paying_users if paying_users > 0 else 0
    
    return {
        'group': group_name,
        'all_users': all_users,
        'paying_users': paying_users,
        'first_deposit_users': first_deposit_users,
        'second_deposit_users': second_deposit_users,
        'total_deposit_sum': total_deposit_sum,
        'key_metric': key_metric,
        'barrier_metric': barrier_metric,
        'app_metric_1': app_metric_1,
        'app_metric_2': app_metric_2
    }

# Считаем метрики для обеих групп
control_metrics = calculate_ab_metrics(abt_df, 'control')
test_metrics = calculate_ab_metrics(abt_df, 'test')

# Создаём сводный DataFrame
metrics_summary = pd.DataFrame([control_metrics, test_metrics])
```

```python
# Анализ динамики

# Функция для вывода метрик по группам
def format_metrics_table(ctrl, test):
    rows = []
    
    metrics_config = [
        ('key_metric', 'Ключевая: ср. сумма на пользователя', 'ден. ед.', 'mean'),
        ('barrier_metric', 'Барьерная: конверсия в 1-й депозит', '%', 'proportion'),
        ('app_metric_1', 'Вспом. 1: конверсия 1→2 депозит', '%', 'proportion'),
        ('app_metric_2', 'Вспом. 2: ср. сумма на платящего', 'ден. ед.', 'mean'),
    ]
    
    for metric_key, label, unit, metric_type in metrics_config:
        ctrl_val = ctrl[metric_key]
        test_val = test[metric_key]
        abs_change = test_val - ctrl_val
        rel_change = (abs_change / ctrl_val * 100) if ctrl_val != 0 else 0
        
        rows.append({
            'Метрика': label,
            'Ед. изм.': unit,
            'Контроль': f'{ctrl_val:,.2f}' if unit == 'ден. ед.' else f'{ctrl_val:.2f}%',
            'Тест': f'{test_val:,.2f}' if unit == 'ден. ед.' else f'{test_val:.2f}%',
            'Абс. изменение': f'{abs_change:+.2f}',
            'Отн. изменение': f'{rel_change:+.2f}%'
            })
    
    return pd.DataFrame(rows)

metrics_table = format_metrics_table(control_metrics, test_metrics)

print(metrics_table)
```

**Формулировка гипотез:**

Для каждой метрики:
 - **H0 (Нулевая)**: Среднее значение метрики в тестовой группе равно среднему в контрольной (μ_test = μ_ctrl).

 - **H1 (Альтернативная)**: Среднее значение метрики в тестовой группе отличается от контрольной (μ_test ≠ μ_ctrl).


```python
#  Функция-замена для z-теста пропорций (работает в любой версии scipy)
def proportions_ztest(success1, nobs1, success2, nobs2):
    p1 = success1 / nobs1
    p2 = success2 / nobs2
    p_pool = (success1 + success2) / (nobs1 + nobs2)  # объединённая пропорция
    se = np.sqrt(p_pool * (1 - p_pool) * (1/nobs1 + 1/nobs2))  # стандартная ошибка
    z_stat = (p1 - p2) / se
    p_value = 2 * norm.sf(abs(z_stat))  # двусторонний p-value
    return z_stat, p_value

results = []

# Барьерная метрика и Вспом.1: Z-test для пропорций
for metric_name, event_filter, desc in [
    ('barrier_metric', 'first_deposit', 'Барьерная: Рег → 1-й депозит'),
    ('app_metric_1', 'second_deposit', 'Вспом. 1: 1-й депозит → 2-й депозит')
]:
    # Подготовка данных
    if metric_name == 'barrier_metric':
        user_events = abt_df.groupby(['user_id', 'group']).apply(
            lambda x: (x['event_name'] == event_filter).any()
        ).reset_index(name='has_event')
    else: 
        users_with_dep1 = abt_df[abt_df['event_name'] == 'first_deposit']['user_id'].unique()
        user_events = abt_df[abt_df['user_id'].isin(users_with_dep1)].groupby(['user_id', 'group']).apply(
            lambda x: (x['event_name'] == event_filter).any()
        ).reset_index(name='has_event')
    
    control_n = len(user_events[(user_events['group'] == 'control')])
    test_n = len(user_events[(user_events['group'] == 'test')])
    
    control_success = user_events[(user_events['group'] == 'control')]['has_event'].sum()
    test_success = user_events[(user_events['group'] == 'test')]['has_event'].sum()
    
    z_stat, p_value = proportions_ztest(test_success, test_n, control_success, control_n)
    
    results.append({
        'metric': metric_name,
        'description': desc,
        'control_rate': control_success/control_n*100,
        'test_rate': test_success/test_n*100,
        'p_value': p_value,
        'significant': p_value < 0.05,
        'test_type': 'z-test (proportions)'
    })
    
    print(f"\n{desc}")
    print(f"   Контроль: {control_success/control_n*100:.2f}% ({control_success}/{control_n})")
    print(f"   Тест:     {test_success/test_n*100:.2f}% ({test_success}/{test_n})")
    print(f"   p-value:  {p_value:.4f} → {'Значимо' if p_value < 0.05 else 'Не значимо'} (α=0.05)")

# Ключевая и Вспом.2: T-test для средних 
for metric_key, desc in [
    ('key_metric', 'Ключевая: ср. сумма на пользователя'),
    ('app_metric_2', 'Вспом. 2: ср. сумма на платящего')
]:
    deposits = abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])]
    user_totals = deposits.groupby(['user_id', 'group'])['amount'].sum().reset_index()
    
    if metric_key == 'key_metric':
        all_users_abt = abt_df.groupby('user_id')['group'].first().reset_index()
        user_data = all_users_abt.merge(user_totals, on=['user_id', 'group'], how='left')
        user_data['amount'] = user_data['amount'].fillna(0)
    else:
        user_data = user_totals[user_totals['amount'] > 0]
    
    control_vals = user_data[user_data['group'] == 'control']['amount']
    test_vals = user_data[user_data['group'] == 'test']['amount']
    
    # Проверка нормальности
    sample_size = min(5000, len(control_vals), len(test_vals))
    _, p_norm_ctrl = stats.shapiro(control_vals.sample(sample_size, random_state=42))
    _, p_norm_test = stats.shapiro(test_vals.sample(sample_size, random_state=42))
    
    # Используем T-Test
    stat, p_value = stats.ttest_ind(test_vals, control_vals)
    
    results.append({
        'metric': metric_key,
        'description': desc,
        'control_mean': control_vals.mean(),
        'test_mean': test_vals.mean(),
        'p_value': p_value,
        'significant': p_value < 0.05,
        'test_type': 'T-Test'
    })
    
    print(f"\n{desc}")
    print(f"   Контроль: {control_vals.mean():,.2f} ({len(control_vals)})")
    print(f"   Тест:     {test_vals.mean():,.2f} ({len(test_vals)})")
    print(f"   Тест:     T-Test")
    print(f"   p-value:  {p_value:.4f} → {'Значимо' if p_value < 0.05 else 'Не значимо'} (α=0.05)")

# Сводная таблица статистики
stats_df = pd.DataFrame(results)
display_cols = ['description', 'p_value', 'significant', 'test_type']
print(stats_df[display_cols].to_string(index=False))
```

```python
# Накопленная КУМУЛЯТИВНАЯ динамика ключевой метрики по дням

# Подготовка: для каждого пользователя — дата регистрации и сумма депозитов
deposits = abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])].copy()
user_totals = deposits.groupby(['user_id', 'group'])['amount'].sum().reset_index()

# Все пользователи с датами регистрации
users_with_dates = abt_df.groupby(['user_id', 'group'])['first_dt'].first().reset_index()
users_with_dates['first_dt'] = pd.to_datetime(users_with_dates['first_dt'])

cohort_data = users_with_dates.merge(user_totals, on=['user_id', 'group'], how='left')
cohort_data['amount'] = cohort_data['amount'].fillna(0)

# Группировка по дням: накопительная средняя
daily_cumulative = []

for group in ['control', 'test']:
    # Берем данные только для текущей группы
    group_data = cohort_data[cohort_data['group'] == group]
    
    # Агрегируем по дням: сумма депозитов и количество пользователей за день
    daily_agg = group_data.groupby('first_dt').agg(
        daily_sum=('amount', 'sum'),
        daily_count=('user_id', 'count')
    ).reset_index().sort_values('first_dt')
    
    # Считаем кумулятивные (накопительные) суммы и количества
    daily_agg['cum_sum'] = daily_agg['daily_sum'].cumsum()
    daily_agg['cum_count'] = daily_agg['daily_count'].cumsum()
    
    # Кумулятивное среднее = накопленная сумма / накопленное количество
    daily_agg['cum_avg'] = daily_agg['cum_sum'] / daily_agg['cum_count']
    
    daily_agg['group'] = group
    daily_cumulative.append(daily_agg)

daily_cum_df = pd.concat(daily_cumulative, ignore_index=True)

# Строим график
plt.figure(figsize=(14, 6))
colors = {'control': '#2E86AB', 'test': '#E74C3C'}

for group in ['control', 'test']:
    group_data = daily_cum_df[daily_cum_df['group'] == group]
    plt.plot(group_data['first_dt'], group_data['cum_avg'], 
             label=f'{group}', color=colors[group], linewidth=2.5, marker='o')

# Финальные значения
final_control = daily_cum_df[daily_cum_df['group'] == 'control']['cum_avg'].iloc[-1]
final_test = daily_cum_df[daily_cum_df['group'] == 'test']['cum_avg'].iloc[-1]

plt.axhline(y=final_control, color=colors['control'], linestyle='--', alpha=0.5)
plt.axhline(y=final_test, color=colors['test'], linestyle='--', alpha=0.5)

plt.title('Кумулятивная динамика ключевой метрики', fontsize=14, fontweight='bold')
plt.xlabel('Дата')
plt.ylabel('Кумулятивная cредняя сумма депозита на пользователя')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()
```

>  **Пояснение к графику кумулятивной динамики**  
По оси X отложены **даты регистрации (когорты пользователей)**, а не дни с момента регистрации (период жизни). 
Кумулятивное среднее пересчитывается накопительно: каждый новый день добавляет свежую когорту, 
что вызывает колебания в начале эксперимента при малой выборке. 
К концу периода линии стабилизируются, отражая устоявшееся значение ключевой метрики (~90–92 ден. ед.).

```python
# Накопленная КУМУЛЯТИВНАЯ динамика ключевой метрики (период жизни)
deposits = abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])].copy()

# Дата депозита
deposits['deposit_date'] = pd.to_datetime(deposits['event_ts']).dt.date

# Дата регистрации для каждого пользователя
daily_revenue = deposits.groupby(['deposit_date', 'group']).agg(
    daily_sum=('amount', 'sum')
).reset_index()

cumulative_by_date = []
for group in ['control', 'test']:
    group_data = daily_revenue[daily_revenue['group'] == group].sort_values('deposit_date').copy()
    
    # Кумулятивная сумма
    group_data['cum_sum'] = group_data['daily_sum'].cumsum()
    
    # Количество пользователей в группе (для расчета средней)
    total_users = abt_df[abt_df['group'] == group]['user_id'].nunique()
    
    # Средняя на пользователя (ключевая метрика)
    group_data['avg_per_user'] = group_data['cum_sum'] / total_users
    
    group_data['group'] = group
    cumulative_by_date.append(group_data)

date_df = pd.concat(cumulative_by_date, ignore_index=True)

# строим график
plt.figure(figsize=(14, 6))
colors = {'control': '#2E86AB', 'test': '#E74C3C'}

for group in ['control', 'test']:
    data = date_df[date_df['group'] == group]
    plt.plot(data['deposit_date'], data['avg_per_user'], 
             label=group, 
             color=colors[group], linewidth=2.5, marker='o')

plt.title('Кумулятивная динамика ключевой метрики', fontsize=14, fontweight='bold')
plt.xlabel('Дата депозита')
plt.ylabel('Накопленная средняя сумма на пользователя')
plt.legend(loc='lower right')
plt.grid(axis='y', alpha=0.5)
plt.xticks()
plt.tight_layout()

plt.show()
```

```python
# Стабильность p-value

# Выбираем активность в event_ts
deposits = abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])].copy()
deposits['act_date'] = pd.to_datetime(deposits['event_ts']).dt.date

# Считаем amount по комбинации (День/пользователь/группа)
daily_user_amounts = deposits.groupby(['act_date', 'user_id', 'group'])['amount'].sum().reset_index()

# Сортируем список уникальных дней активности
act_dates = sorted(daily_user_amounts['act_date'].unique())

# Добавим справочники для быстрого доступа
user_groups = abt_df.groupby('user_id')['group'].first().to_dict()
user_reg_dates = abt_df.groupby('user_id')['first_dt'].first().dt.date.to_dict()

pvalues_over_time = []
dates_for_plot = []

# Накопительный расчёт ARPU по дням
for date in act_dates:
    
    # Фильтруем данные до этой даты
    cum_data = daily_user_amounts[daily_user_amounts['act_date'] <= date]
    
    # Считаем ARPU накопительно
    user_arpu = cum_data.groupby(['user_id', 'group'])['amount'].sum().reset_index(name='cum_arpu')
    
    # Фильтруем клиентов, только уже зарегистрированых
    valid_users = [uid for uid, reg_d in user_reg_dates.items() if reg_d <= date]                         
    pool_df = pd.DataFrame({'user_id': valid_users})
    pool_df['group'] = pool_df['user_id'].map(user_groups)
    
    pool_df = pool_df.merge(user_arpu, on=['user_id', 'group'], how='left')
    pool_df['cum_arpu'] = pool_df['cum_arpu'].fillna(0)                                
                                    
    # Делим по группам
    ctrl = pool_df[pool_df['group'] == 'control']['cum_arpu']
    test = pool_df[pool_df['group'] == 'test']['cum_arpu']
    
    # Пропускаем дни с маленькой выборкой
    if len(ctrl) < 30 or len(test) < 30:
        pvalues_over_time.append(np.nan)
        dates_for_plot.append(date)
        continue
    
    # T-Test
    stat, p_val = stats.ttest_ind(test, ctrl)
    pvalues_over_time.append(p_val)
    dates_for_plot.append(date)


plt.figure(figsize=(12, 5))
plt.plot(dates_for_plot, pvalues_over_time, marker='o', linestyle='-', color='#2E86AB', linewidth=2)
plt.axhline(y=0.05, color='red', linestyle='--', label='Порог значимости (α=0.05)', linewidth=1.5)
plt.title('Динамика p-value ключевой метрики (накопительно по дням)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Дата регистрации')
plt.ylabel('p-value')
plt.legend()
plt.grid(alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

Новая фича могла повлиять на поведение пользователей.
- Пользователи, которые раньше вносили небольшие суммы, могли стать более осторожными, сократить свои вложения или совсем перестать платить. Это отразится в снижении 25-го перцентиля суммы депозитов в тестовой группе.
- Пользователи, которые склонны к более крупным инвестициям, могли сильнее вовлечься в продукт и начать вносить больше средств. Это отразится в росте 75-го перцентиля.

Используя бутстрап, сравните разницы перцентилей (25, 50, 75) суммы всех депозитов на платящего пользователя в контрольной и тестовой группах:
- Постройте доверительные интервалы для разницы между группами для каждого перцентиля.
- Интерпретируйте изменения в контексте сформулированной выше гипотезы.
- Визуализируйте результат бутстрап-теста. 

```python
# Фиксируем сид для воспроизводимости
np.random.seed(42)
```

```python
# Подготовка данных

# Фильтруем только события депозитов
deposits = abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])].copy()

# Суммируем все депозиты для каждого пользователя
user_totals = deposits.groupby(['user_id', 'group'])['amount'].sum().reset_index()

# Оставляем только платящих пользователей (сумма > 0)
paying_users = user_totals[user_totals['amount'] > 0].copy()

# Разделяем по группам для удобства
control_amounts = paying_users[paying_users['group'] == 'control']['amount'].values
test_amounts = paying_users[paying_users['group'] == 'test']['amount'].values

# Базовая статистика
print(f"\nСтатистика сумм депозитов:")
for name, data in [('Контроль', control_amounts), ('Тест', test_amounts)]:
    print(f"\n{name}:")
    print(f"   25-й перцентиль: {np.percentile(data, 25):.2f}")
    print(f"   50-й перцентиль (медиана): {np.percentile(data, 50):.2f}")
    print(f"   75-й перцентиль: {np.percentile(data, 75):.2f}")
    print(f"   Среднее: {data.mean():.2f}, Макс: {data.max():.2f}")
```

```python
# Функция бутстрапа для разницы перцентилей

def bootstrap_percentile_diff(group1, group2, percentile, n_iterations=10000, random_state=42):
    np.random.seed(random_state)
    
    # Наблюдаемые перцентили
    p_ctrl = np.percentile(group1, percentile)
    p_test = np.percentile(group2, percentile)
    diff_observed = p_test - p_ctrl
    
    # Бутстрап
    boot_diffs = []
    for _ in range(n_iterations):
        # Сэмплируем с заменой
        boot_ctrl = np.random.choice(group1, size=len(group1), replace=True)
        boot_test = np.random.choice(group2, size=len(group2), replace=True)
        
        # Считаем перцентили и разницу
        p_boot_ctrl = np.percentile(boot_ctrl, percentile)
        p_boot_test = np.percentile(boot_test, percentile)
        boot_diffs.append(p_boot_test - p_boot_ctrl)
    
    boot_diffs = np.array(boot_diffs)
    
    # 95% доверительный интервал
    ci_lower = np.percentile(boot_diffs, 2.5)
    ci_upper = np.percentile(boot_diffs, 97.5)
    
    # Значимость: ДИ не включает 0
    significant = (ci_lower > 0) or (ci_upper < 0)
    
    return {
        'percentile': percentile,
        'p_ctrl': p_ctrl,
        'p_test': p_test,
        'diff_observed': diff_observed,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper,
        'boot_diffs': boot_diffs,
        'significant': significant
    }

results = {}
percentiles = [25, 50, 75]

for p in percentiles:
    result = bootstrap_percentile_diff(control_amounts, test_amounts, percentile=p, n_iterations=10000)
    results[p] = result
    
    sig_marker = "Значимо" if result['significant'] else "Не значимо"
    direction = "рост" if result['diff_observed'] > 0 else "падение"
    
    print(f"\n{p}-й перцентиль:")
    print(f"   Контроль: {result['p_ctrl']:.2f}")
    print(f"   Тест:     {result['p_test']:.2f}")
    print(f"   Разница:  {result['diff_observed']:+.2f} {direction}")
    print(f"   95% ДИ:   [{result['ci_lower']:+.2f}; {result['ci_upper']:+.2f}]")
    print(f"   Статус:   {sig_marker}")
```

```python
# строим график

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Бутстрап-распределение разницы перцентилей (Тест - Контроль)', fontsize=14, fontweight='bold')

colors = {True: '#2ECC71', False: '#95A5A6'}  # зелёный для значимых, серый для незначимых

for idx, (p, ax) in enumerate(zip(percentiles, axes)):
    res = results[p]
    
    # Гистограмма бутстрап-разниц
    ax.hist(res['boot_diffs'], bins=50, alpha=0.7)
    
    # Линии доверительного интервала
    ax.axvline(res['ci_lower'], color='red', linestyle='--', linewidth=2, label='95% ДИ')
    ax.axvline(res['ci_upper'], color='red', linestyle='--', linewidth=2)
    
    
    # Заголовок и подписи
    ax.set_xlabel('Разница перцентиля (Тест - Контроль)')
    ax.set_ylabel('Частота')
    ax.legend(fontsize=9)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()

plt.show()
```
